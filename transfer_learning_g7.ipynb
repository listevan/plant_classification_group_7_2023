```python
# import torch and numpy and pretrained model
from torchvision import models
import torch.nn as nn
import torch
import numpy as np
import os

# load pretrained model
pretrained_model = models.resnet50(weights = 'DEFAULT', progress = True)
```

```python
for (name, layer) in pretrained_model._modules.items():
    print(name, " : ", layer)
```

```python
for param in pretrained_model.parameters():
    param.requires_grad = False

res = list(pretrained_model.children())[:-2]
model = nn.Sequential(*res)
```

```python
for (name, layer) in model._modules.items():
    print(name, " : ", layer)
```

```python
relu = nn.ReLU()
dropout = nn.Dropout(p=0.2)
softmax = nn.Softmax(15)

model.extend([relu, dropout, softmax])
```

```python
device = (
    "cuda"
    if torch.cuda.is_available()
    else "mps"
    if torch.backends.mps.is_available()
    else "cpu"
)
print(f"Using {device} device")
```
